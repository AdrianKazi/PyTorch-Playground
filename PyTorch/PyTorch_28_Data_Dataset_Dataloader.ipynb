```python
import numpy as np
import torch
from torch.utils.data import DataLoader, TensorDataset
```

# Datasets

```python
nObservations = 100
nFeatures = 20

data = np.random.randn(nObservations, nFeatures)
data.shape
```

```python
# Convert to torch tensor
dataT = torch.tensor(data)

# print info
print('Numpy data:')
print(type(data))
print(data.shape)
print(data.dtype)
print(' ')

print('Tensor data:')
print(type(dataT))
print(dataT.size() )
print(dataT.dtype)
```

```python
# COnverting data types

dataT2 = torch.tensor(data).float()
print(dataT2.dtype)

dataT3 = torch.tensor(data).long()
print(dataT3.dtype)
```

```python
# Dataset Tensors
dataset = TensorDataset(dataT)
dataset.tensors
```

```python
# With Labels
labels = torch.ceil(torch.linspace(.01, 4, nObservations))

# trainsorm to an actual matrix (column vactor)
labels = labels.reshape((-1,1))

dataset = TensorDataset(dataT, labels)
print(dataset.tensors[0].size())
print(dataset.tensors[1].size())

print(np.shape(np.random.randint(5, size = nObservations)))
```

# DataLoaders

```python
batch_size = 25
dataloader = DataLoader(dataset, batch_size = batch_size, shuffle = False) # drop_last = True
```

```python
# size of each batch
for data, labs in dataloader:
  print('BATCH INFO:')
  print(data.size())
  print(labs.size())
  print(' ')
```

```python
# size of each batch
for data, labs in dataloader:
  print(labs.T)
  print(' ')
```

```python
# with shuffle
dataloader = DataLoader(dataset, batch_size = batch_size, shuffle = True)
for data, labs in dataloader:
  print(labs.T)
  print(' ')
```
