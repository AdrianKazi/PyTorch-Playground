# Torchvision

```python
import torchvision
import pickle
import numpy as np
import matplotlib.pyplot as plt

# download the CIFAR10 dataset
cdata = torchvision.datasets.CIFAR10(root = 'cifar10', download = True)
cdata
```

```python
with open('/content/cifar10/cifar-10-batches-py/data_batch_1', 'rb') as f:
  batch = pickle.load(f, encoding='bytes')

X = batch[b'data']
y = batch[b'labels']

# reshape data to (num_images, height, width, channels)
X = X.reshape(-1, 3, 32,32).transpose(0,2,3,1)

print(X.shape, len(y))
```

```python
plt.imshow(X[np.random.randint(X.shape[0])])
plt.show()
```
